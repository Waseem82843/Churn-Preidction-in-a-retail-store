```python
#Historic Visits vs CHURN
plt.figure(figsize=(15,10))
ax = sns.countplot(x='HISTORIC_VISIT', hue='CHURN', data=df);
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

```python
df["dif"]=df["FIRST_VISIT_DAY"]-df["LAST_VISIT_DAY"]
```

```python
df.head()
```

```python
plt.figure(figsize=(15,10))
ax = sns.countplot(x='dif', hue='CHURN', data=df);
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```
